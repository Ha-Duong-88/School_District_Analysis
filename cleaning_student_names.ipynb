```python
# Module 4.5.4 - Get the Incorrect Student Names
# To get all the students' names in a separate file, we'll use the Pandas tolist() method on a DataFrame with reference to the student_name column. 
# Using the tolist() method on the student_name column will add all the names to a list.
# Recall that every column in the CSV file and DataFrame is a Series or a list.
# Sudent names is a column of student_data_df DataFrame
# Put the student names in a list.
student_names = student_data_df["student_name"].tolist()
student_names
```

```python
# Module 4.5.4 - The split() Method
# The split() method will split a Python string object on the whitespace, or where there is no text.
# If we split each name, we can count the length of each name. 
# A student with a first and last name only will return a length of 2, whereas a name with a prefix or suffix only will have a length of 3.
# Use split() method with for loop to iterate
```

```python
# Module 4.5.4 - The split() Method
# NOTE - We put all the student names in a list above
# In the student_names list, we can use a for loop to iterate through the list, split each name, 
# and print out the name and the length of the name.
# Define student_names.pd first (I added this to resolve error)
# Split the student name and determine the length of the split name.

student_names = student_data_df["student_name"]

for name in student_names:
    print(name.split(), len(name.split()))
```

```python
# Module 4.5.4 - Conditional statement
# We can filter this list using a conditional statement. If the length of the name is equal to or greater than 3, 
# we can add it to a new list.
#  By getting the length of this new list, we can determine how many students are affected.
# Create a new list and use it for the for loop to iterate through the list.
students_to_fix = []

# Use an if statement to check the length of the name.
# If the name is greater than or equal to "3", add the name to the list.

for name in student_names:
    if len(name.split()) >= 3:
        students_to_fix.append(name)

# Get the length of the students whose names are greater than or equal to "3".
len(students_to_fix)

```

```python
# Module 4.5.4 - Conditional statement
# cont'd from above to print the students_to_fix
print(students_to_fix)
```

```python
# Module 4.5.4 - Conditional statement
# Module 4.5.4 - Get All the Prefixes
# To get all of the prefixes in the student data, we're going to write an algorithm that will iterate through the students_to_fix list
# and determine if the length of the prefix, or the first item in the list after using name.split(), is less than or equal to 4. 
# We'll use 4 as the upper limit because the longest prefix is "Miss," which is four characters long.
# If the length is less than or equal to 4, we'll add it to a new list named prefixes.
# Add the prefixes less than or equal to 4 to a new list.
prefixes = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])

print(prefixes)
```

```python
# Module 4.5.4 - Get All the Suffixes
# Let's apply the same methodology to get the suffixes. We'll iterate through the students_to_fix list and determine 
# if the last item [-1] after using name.split() is less than or equal to 3.
# We can use negative indexing to get the last item in the list.
# Add the suffixes less than or equal to 3 to a new list.
suffixes = []
for name in students_to_fix:
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])

print(suffixes)
```

```python
# Module 4.5.4 - The set() Method
# The set() method returns all unique items in a list when that list is added inside the parentheses.
# If the list contains strings, they will be ordered alphabetically.
# Get the unique items in the "prefixes" list.
set(prefixes)
```

```python
# Module 4.5.4 - The set() Method
# Now let's apply the same method to the suffixes.
# The set() method returns all unique items in a list when that list is added inside the parentheses.
# If the list contains strings, they will be ordered alphabetically.
# Get the unique items in the "suffixes" list.
set(suffixes)

# Now we can remove Dr., Mr., Mrs., Ms., Miss, MD, DDS, DVM, and Ph.D. from the students' names in student_data_df.
```

```python
# Module 4.5.5 - Remove Strings: strip() vs. replace()
# The strip() Method
# The strip() method removes any combination of letters and words that are inside the parentheses
# For example, if we put "Mrs." inside the parentheses of the strip() method, any combination of the letters "M," "r," 
# and "s" will be stripped from the student names.
# Strip "Mrs." from the student names
for name in students_to_fix:
    print(name.strip("Mrs."))
```

```python
# Module 4.5.5 - Remove Strings: strip() vs. replace()
# The replace() Method
# When we use the replace() method, we can replace an "old" phrase, or string, with a new phrase, or string, inside 
# the parentheses.
# In this example, we want to replace prefixes with an empty string, or "nothing."
# Replace "Dr." with an empty string.
name = "Dr. Linda Santiago"
#name.replace("Dr.", "")


# Notice in the output that there is a white space before "Linda Santiago." We can remove this white space by adding it to "Dr." while we are using the replace() method, like this:
name.replace("Dr.", "")

# To change "Mrs. Linda Santiago" to "Dr. Linda Santiago" you would write 
# name.replace("Mrs.","Dr.")
```

```python
# 4.5.6 - Replace substrings
# load and read student_data
student_data_to_load = "Resources/students_complete.csv"
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
# 4.5.6 - Replace substrings
# We can use the replace() method to remove prefixes and suffixes from the students' names.
# However, we don't want to write a different replace() statement for each prefix and suffix;
# Better way to do it is
# Step 1 - Declare a list named prefixes_suffixes that holds the common prefixes and suffixes as strings.
# Step 2 - Iterate through this list with a for loop.
# Step 3 - For each item in the list, use the replace() method on the student_name column.
# Make sure to first load the student and read student data file (see above) first else 
# below code will get a NameError

# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]


# Remember that there is a whitespace between the prefix and the first name,
# We're going to add a whitespace after each prefix and before each suffix in our list of prefixes and suffixes.
# replace() method can be used only on Python strings.
# use dtypes method to determine the data types of each column in a DataFrame. For the student_data_df DataFrame


# student_name is a column in the student_dta_df DataFrame
# pass suffix or prefix as word in the replace() method
# refer to module for breakdown of what is happening in the code
# Iterate through the "prefixes_suffixes" list and replace them with an empty space, "" when it appears in the student's name.
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
    
# Print the first 10 rows of student_data_df
student_data_df
```

```python
# Module 4.5.6 - Replace Substrings 
# cont'd from above code block
# Now we have to confirm that all of the prefixes and suffixes were caught in our script (the code that written above)
# Step 1 - Declare a new variable and assign it to the student_data_df["student_name"]
# Step 2 - Convert the student_data_df["student_name"] to a list using .tolist() like we did in
# previous module
# Put the cleaned students' names in another list.
student_names = student_data_df["student_name"].tolist()
student_names
```

```python
# Module 4.5.6 - Replace Substrings 
# cont'd from above code block to put the cleaned
# student's names in another list
# Now, add all the students' names that are greater than or equal to 3 to a list.
# Create a new list and use it for the for loop to iterate through the list.
students_fixed = []

# Use an if statement to check the length of the name.

# If the name is greater than or equal to 3, add the name to the list.
for name in student_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)

# Get the length of the students' names that are greater than or equal to 3.
len(students_fixed)

# NOTE on output: There are now only names with familial designations like Jr., II, III, IV, and V in the students_fixed list, 
# which confirms that we caught all the professional prefixes and suffixes
```
