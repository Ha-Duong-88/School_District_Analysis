```python
# 4.5.6 
# Replace sub strings
# A more efficient way to remove the prefixes and suffixes is to do the following instead of for each prefix and suffic
# adding each prefix and suffix to remove to a list.
# Declare a list named prefixes_suffixes that holds the common prefixes and suffixes as strings.
# For each item in the list, use the replace() method on the student_name column.
# Iterate through this list with a for loop.
# we need to convert the object data type to a string while using the replace() method. 
# For each item in the list, use the replace() method on the student_name column.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
student_data_df["student_name"].str.replace()
# Iterate through the "prefixes_suffixes" list and replace them with an empty space, "" when it appears in the student's name.
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
```
